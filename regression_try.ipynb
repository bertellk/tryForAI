```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/laptop_price - dataset.csv", encoding="ISO-8859-1")
#X = dataset.iloc[:, :-1].values
#y = dataset.iloc[:, -1].values
selected_data = dataset[['CPU_Frequency (GHz)', 'RAM (GB)','OpSys','Weight (kg)', 'Price (Euro)']]
```

```python
print(dataset.columns)
```

```python
X = selected_data.iloc[:, :-1].values
y = selected_data.iloc[:, -1].values

"""selected_data = dataset[['Memory']].copy()
selected_data['Memory'] = selected_data['Memory'].astype(str)

selected_data = dataset[['GPU_Type']].copy()
selected_data['GPU_Type'] = selected_data['GPU_Type'].astype(str)

selected_data = dataset[['OpSys']].copy()  # Alt kümeyi kopyala
selected_data['OpSys'] = selected_data['OpSys'].astype(str)  # Dönüşüm işlemi"""
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[2] )], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
